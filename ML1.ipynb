```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler
```

```python
X_train,X_test,y_train,y_test= train_test_split(X,Y,test_size=.30,random_state=100)
```

```python
X_train
```

# encoding

```python
encoder=OneHotEncoder(handle_unknown='ignore')
```

```python
X_train_encoded = encoder.fit_transform(X_train[['Caption','Hashtags']])
X_train_encoded_df=pd.DataFrame(X_train_encoded.toarray())
X_train_encoded_df

```

```python
X_test_encoded =encoder.transform(X_test[['Caption','Hashtags']])
X_test_encoded_df=pd.DataFrame(X_test_encoded.toarray())
```

# scaling

```python
scalar=StandardScaler()
```

```python
X_train_scalar = scalar.fit_transform(X_train[['Followers','Time']])
```

```python
X_test_scalar = scalar.transform(X_test[['Followers','Time']])
```

```python
X_test_scalar_arr=np.array(X_test_scalar)
X_train_scalar_arr=np.array(X_train_scalar)
```

```python
X_train_df1=pd.DataFrame(data=X_train_scalar_arr,columns=['Followers','Time'])
X_train_df1
```
